```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt

import statsmodels.formula.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

Treina a regressao linear com o cojunto de treino

```python
treino = pd.read_csv("treino.csv")
treino = pd.DataFrame(treino,columns=treino.columns[1:])
treino
```

```python
fixed_acidity = treino.iloc[:,0].values
volatile_acidity = treino.iloc[:,1].values
citric_acid = treino.iloc[:,2].values
residual_sugar = treino.iloc[:,3].values
chlorides = treino.iloc[:,4].values
free_sulfur = treino.iloc[:,5].values
total_sulfur = treino.iloc[:,6].values
density = treino.iloc[:,7].values
ph = treino.iloc[:,8].values
sulphates = treino.iloc[:,9].values
alcohol  = treino.iloc[:,10].values
```

```python
import statsmodels.formula.api as sm
modeloFinalajustado = sm.ols(data=treino, formula="quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur + total_sulfur + density + pH + sulphates + alcohol")
modeloFinaltreinado = modeloFinalajustado.fit()
modeloFinaltreinado.summary()
```

Testando a regressão treinada

```python
teste = pd.read_csv("teste.csv")

teste = pd.DataFrame(teste,columns=teste.columns[1:])
teste_quality = teste.iloc[:,-1].values

#Gambiarra
aux = pd.DataFrame({'fixed acidity': 0, 'volatile acidity': [0], 'citric acid': [0], 'residual sugar': [0], 'chlorides': [0], 'free sulfur': [0], 'total sulfur': [0], 'density': [0], 'ph': [0], 'sulphates': [0], 'alcohol': [0], 'quality': [0]})
for i in range(len(treino)-len(teste)):
    teste = pd.concat([teste,aux],ignore_index=True,axis=0)

teste
```

```python
pred = modeloFinaltreinado.predict(exog=teste)
aux = pred.to_numpy()
pred = [*aux[:480]]
print (pred)
```

Raiz Quadrada da Média dos erros

```python
rms = sqrt(mean_squared_error(pred, teste_quality))
print (rms)
```
