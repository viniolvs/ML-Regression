```python
import sklearn
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

Define as bases de treino e teste

```python
teste = pd.read_csv("teste.csv")
teste_variaveis = teste.iloc[:,1:11]
teste_quality = teste.iloc[:,-1]

treino = pd.read_csv("treino.csv")
treino_variaveis = treino.iloc[:,1:11]
treino_quality = treino.iloc[:,-1]
treino
```

```python
teste
```

MLP regressor

```python
#treina com a base de treino
regr = MLPRegressor(hidden_layer_sizes=(5,5,5),max_iter=500).fit(treino_variaveis, treino_quality)
#testa com a base de teste
pred_mlp = regr.predict(teste_variaveis)
print(*pred_mlp,sep=' | ')
```

SVR

```python
svr = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1).fit(treino_variaveis,treino_quality)
pred_svr = svr.predict(teste_variaveis)
print(*pred_svr,sep=' | ')
```

Random Forest Regressor

```python
RFR = RandomForestRegressor(max_depth=2, random_state=0).fit(treino_variaveis,treino_quality)
pred_rfr = RFR.predict(teste_variaveis)
print(*pred_rfr, sep=" | ")
```

Avaliando os Regressores

MLP regressor

```python
#Raiz do erro quadrático médio
rmse = mean_squared_error(pred_mlp, teste_quality,squared=True)
print (rmse)

#Erro Quadrático Médio
mse = mean_squared_error(pred_mlp, teste_quality,squared=False)
print (mse)
```

SVR

```python
#Raiz do erro quadrático médio
rmse = mean_squared_error(pred_svr, teste_quality,squared=True)
print (rmse)

#Erro Quadrático Médio
mse = mean_squared_error(pred_svr, teste_quality,squared=False)
print (mse)
```

Random Forest Regressor

```python
#Raiz do erro quadrático médio
rmse = mean_squared_error(pred_rfr, teste_quality,squared=True)
print (rmse)

#Erro Quadrático Médio
mse = mean_squared_error(pred_rfr, teste_quality,squared=False)
print (mse)
```
