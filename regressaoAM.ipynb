```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

Define as bases de treino e teste

```python
teste = pd.read_csv("teste.csv")
teste_variaveis = teste.iloc[:,1:11]
teste_quality = teste.iloc[:,-1]

treino = pd.read_csv("treino.csv")
treino_variaveis = treino.iloc[:,1:11]
treino_quality = treino.iloc[:,-1]
treino
```

```python
teste
```

MLP regressor

```python
#Define os melhores parÃ¢metros utilizando a base de treino
erros = []
param = []

for i in range (10):
    for j in range (10):
        mlp = MLPRegressor(hidden_layer_sizes=(i+1,i+1,i+1),max_iter=(j+1)*50).fit(treino_variaveis, treino_quality)
        pred_mlp = mlp.predict(treino_variaveis)
        rmse = mean_squared_error(pred_mlp, treino_quality,squared=True)
        erros.append(rmse)
        param.append([i+1,(j+1)*50])

best_mlp = np.argmin(erros)
layers = param[best_mlp][0]
max_iter = param[best_mlp][1]

print(best_mlp,layers,max_iter)
```

```python
#treina com os melhores parÃ¢metros
mlp = MLPRegressor(hidden_layer_sizes=(layers,layers,layers),max_iter=max_iter).fit(treino_variaveis, treino_quality)
#testa com a base de teste
pred_mlp = mlp.predict(teste_variaveis)

#Avalia
print("MLP") 
#Raiz do erro quadrÃ¡tico mÃ©dio
rmse = mean_squared_error(pred_mlp, teste_quality,squared=True)
print ("Raiz do erro quadrÃ¡tico mÃ©dio =",rmse)

#Erro QuadrÃ¡tico MÃ©dio
mse = mean_squared_error(pred_mlp, teste_quality,squared=False)
print ("Erro QuadrÃ¡tico MÃ©dio =",mse)
```

SVR

```python
#Define os melhores parÃ¢metros

#kernel radial
erros = []
param = []

for i in range (10):
    for j in range (10):
        svr = SVR(kernel='rbf', C=(i+1)*10, gamma=(j+1)/10, epsilon=(j+1)/10).fit(treino_variaveis,treino_quality)
        pred = svr.predict(treino_variaveis)
        rmse = mean_squared_error(pred, treino_quality,squared=True)
        erros.append(rmse)
        param.append([(i+1)*10,(j+1)/10])

best_svr = np.argmin(erros)

C = param[best_svr][0]
gamma = param[best_svr][1]
epsilon = param[best_svr][1]

print(C,gamma,epsilon)
```

```python
#treina com os melhores parÃ¢metros
svr = SVR(kernel='rbf', C=C, gamma=gamma, epsilon=epsilon).fit(treino_variaveis,treino_quality)
pred_svr = svr.predict(teste_variaveis)

#Avalia
print("SVR")
#Raiz do erro quadrÃ¡tico mÃ©dio
rmse = mean_squared_error(pred_svr, teste_quality,squared=True)
print ("Raiz do erro quadrÃ¡tico mÃ©dio =",rmse)

#Erro QuadrÃ¡tico MÃ©dio
mse = mean_squared_error(pred_svr, teste_quality,squared=False)
print ("Erro QuadrÃ¡tico MÃ©dio =",mse)
```

Random Forest Regressor

```python
#Define os melhores parÃ¢metros
erros = []
param = []

for i in range (7):
    for j in range (7):
        RFR = RandomForestRegressor(max_depth=i+1, random_state=j+1).fit(treino_variaveis,treino_quality)
        pred = RFR.predict(treino_variaveis)
        rmse = mean_squared_error(pred, treino_quality,squared=True)
        erros.append(rmse)
        param.append([i+1,j+1])

best_rfr = np.argmin(erros)
depth = param[best_rfr][0]
random = param[best_rfr][1]

print(depth,random)
```

```python
#treina com os melhores parÃ¢metros
rfr = RandomForestRegressor(max_depth=depth, random_state=random).fit(treino_variaveis,treino_quality)
pred_rfr = rfr.predict(teste_variaveis)

#Avalia
print("RFR")
#Raiz do erro quadrÃ¡tico mÃ©dio
rmse = mean_squared_error(pred_rfr, teste_quality,squared=True)
print ("Raiz do erro quadrÃ¡tico mÃ©dio =",rmse)

#Erro QuadrÃ¡tico MÃ©dio
mse = mean_squared_error(pred_rfr, teste_quality,squared=False)
print ("Erro QuadrÃ¡tico MÃ©dio =",mse)
```
